```python
import glob
import netCDF4 as nc
import pandas as pd
import numpy as np
import xarray as xr
import xesmf as xe
import cartopy.crs as ccrs
import os
```

```python
for model in ['HadGEM2-ES', 'MIROC-ESM', 'CanESM2']:
    files_list = glob.glob(rf"/home/uditbhatia/Documents/Sarth/Downscaling_AugmentedConvLSTM/02_Raw_Data/CMIP5 Data/pr_day_{model}_historical_r1i1p1_*.nc")
    print(f"{model}: {len(files_list)}")
    ds = xr.open_mfdataset(files_list,combine = 'by_coords')
    print(ds.dims)
    print(ds.coords)
    print()
    ds = ds.sel(time=((ds.time.dt.year >= 1948) & (ds.time.dt.year <= 2005)))
    print(ds.dims)
    print(ds.coords)
    print()
    ds = ds.sel(time=~((ds.time.dt.month == 2) & (ds.time.dt.day == 29)))
    ds = ds.sel(time=~((ds.time.dt.month == 2) & (ds.time.dt.day == 30)))
    lat_bnds, lon_bnds = [6, 39], [66, 100.5]
    ds = ds.sel(lat=slice(*lat_bnds), lon=slice(*lon_bnds))
    print(ds.dims)
    print(ds.coords)
    print()
    ds.to_netcdf(rf"/home/uditbhatia/Documents/Sarth/Downscaling_AugmentedConvLSTM/03_Preprocess_Data/NC_Files/{model}.nc")
    print("****************************************************************************")
    print()
```

# MIROC-ESM

```python
ds = xr.open_dataset(r"/home/uditbhatia/Documents/Sarth/Downscaling_AugmentedConvLSTM/03_Preprocess_Data/NC_Files/MIROC-ESM.nc")
ds
```

```python
lat_bnds, lon_bnds = [6.5, 38.5], [66.5, 100]
projection_dimension = [129, 135]
lat_lst = np.linspace(lat_bnds[0], lat_bnds[1], projection_dimension[0])
lon_lst = np.linspace(lon_bnds[0], lon_bnds[1], projection_dimension[1])
```

```python
dr = ds["pr"]
dr
```

```python
ds_out = xr.Dataset(
    {
        "lat": (["lat"], lat_lst),
        "lon": (["lon"], lon_lst),
    }
)
ds_out
```

```python
regridder = xe.Regridder(ds, ds_out, "bilinear")
regridder  # print basic regridder information.
```

```python
dr_out = regridder(dr)
dr_out
```

```python
dr_np = dr_out.to_numpy()
dr_np.shape
```

```python
np.save(
    r"/home/uditbhatia/Documents/Sarth/Downscaling_AugmentedConvLSTM/03_Preprocess_Data/npy_Files/MIROC-ESM.npy", 
    dr_np
    )
```
